## Data PreProcessing :

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df = pd.get_dummies(df, columns=['Education', 'City', 'Gender', 'EverBenched'], drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['Age', 'ExperienceInCurrentDomain', 'JoiningYear', 'PaymentTier']

df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('LeaveOrNot', axis=1)
y = df['LeaveOrNot']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=101
)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=101)

dt_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_dt = dt_model.predict(x_test)

accuracy_dt = accuracy_score(y_test, y_pred_dt)

print(" Decision Tree - The Accuracy Score is:",
      str(round(accuracy_dt * 100, 2)) + "%")
```
